繪製一個圓心在原點，半徑為 1 的圓的函數圖形

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#產生一個包含400個數值的等距數列，範圍從 -1.2 到 1.2，x 和 y 分別代表x和y軸上的點。
x = np.linspace(-1.2, 1.2, 400)
y = np.linspace(-1.2, 1.2, 400)

#建立2D網格
X, Y = np.meshgrid(x, y)

#圓的方程式
Z = X**2 + Y**2 - 1

#建立圖形區域，並設定大小為6×6英吋
plt.figure(figsize=(6,6))

#繪製圓的邊界(Z=0)
plt.contour(X, Y, Z, levels=[0], colors='blue')

#畫x軸
plt.axhline(0, color='black', linewidth=0.5)

#畫y軸
plt.axvline(0, color='black', linewidth=0.5)

#設定x、y軸的顯示範圍
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)

#圖形標題
plt.title("Circle")

#網格線
plt.grid(True, linestyle="--", alpha=0.6)

#顯示圖表
plt.show()
```
