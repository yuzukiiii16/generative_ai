**目標：** 使用者輸入問題，根據我整理的LLM prompt框架生成範例並引導學生思考如何下prompt，讓學生可以透過這個prompt提升與LLM互動的效率

#### ✅ CoT 流程：
1. **第一階段（思考階段）**：請 LLM 根據使用者提供的問題，進行以下思考流程
* 步驟一：請針對以下五個要素逐一進行檢查。對每個要素，請依序回答以下兩點：
    - 檢查結果：說明原始提問是否已涵蓋該要素，若描述不清楚，請指出問題。
    - 補充建議：給出具體的改善方向與補充內容建議。
    
    五個要素如下：
    1. 主題（Theme）：是否清楚描述正在解決的問題主題或情境背景？
    2. 目標（Aim）：是否明確指出希望 AI 協助達成的具體目標，並說明使用的工具、方法或語言？
    3. 角色（Persona）：是否指定 AI 扮演特定角色（如資訊科學老師、資料科學家、工程師、資工系教授）？
    4. 對象（Recipients）：是否交代輸出內容的目標讀者或使用者（如學生、教師、開發者）？
    5. 範本（Template）：是否提供回應格式、語言風格或其他輸出限制條件？

* 步驟二：請檢查原始提問是否符合以下三項 prompt 設計原則，並對每項原則分別回答以下兩點：
    - 檢查結果：說明是否符合原則，若不符合請指出問題。
    - 補充建議：提供具體修改方向以提升該原則的符合度。

    三項設計原則如下：
    1. 簡潔（Concise）：語句是否清楚、是否去除冗詞？
    2. 有邏輯（Logical）：內容結構是否合理、順序是否清晰？
    3. 明確（Explicit）：是否明白交代所需輸出內容與格式？

2. **第二階段（產出階段）**：請 LLM 將上面思考整合為一個prompt建議與5個範例，讓學生可以參考，並且須符合以下要求：
* 不可以和學生說答案，只能提供prompt的範例
* 給予提問建議
* 提供5個範例(會參考RAG的內容)

```python
class CustomE5Embedding(HuggingFaceEmbeddings):
    def embed_documents(self, texts):
        texts = [f"passage: {t}" for t in texts]
        return super().embed_documents(texts)

    def embed_query(self, text):
        return super().embed_query(f"query: {text}")
```

```python
# Define the LLM using Groq
llm = ChatGroq(
    api_key=os.environ["GROQ_API_KEY"],
    model="gemma2-9b-it"
)

# Create the prompt template with system and human messages
prompt = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template(system_prompt),
    HumanMessagePromptTemplate.from_template(prompt_template)
])

# Create the LangChain chain
chain = (
    {"retrieved_chunks": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

# ... (your reply function and Gradio interface)

def reply(user_input,
          provider="groq",
          model="gemma2-9b-it"):

    global chat_history

    # Call the chain with the user input
    answer = chain.invoke({"question": user_input})

    # Format the user message and AI response as dictionaries
    user_message = {"role": "user", "content": user_input}
    ai_response = {"role": "assistant", "content": answer}

    # Append the messages as dictionaries to chat_history
    chat_history.append(user_message)
    chat_history.append(ai_response)

    # ✅ Add print statements to inspect chat_history
    print("--- Current chat_history before returning ---")
    for message in chat_history:
        print(message)
    print("-------------------------------------------")


    # Return the chat history in the correct format
    return chat_history

```

```python
def Question(prompt):
    global chat_history
    # 取回相關資料
    docs = retriever.get_relevant_documents(user_input)
    retrieved_chunks = "\n\n".join([doc.page_content for doc in docs])

    # Step 1: CoT - 思考三步驟
    planning_prompt = f"""
    使用者輸入的提問是：{prompt}

    請依照以下兩個步驟，進行具體分析與補充建議：

    步驟一：請針對以下五個要素逐一進行檢查。對每個要素，請依序回答以下兩點：
    - 檢查結果：說明原始提問是否已涵蓋該要素，若描述不清楚，請指出問題。
    - 補充建議：給出具體的改善方向與補充內容建議。

    五個要素如下：
    1. 主題（Theme）：是否清楚描述正在解決的問題主題或情境背景？
    2. 目標（Aim）：是否明確指出希望 AI 協助達成的具體目標，並說明使用的工具、方法或語言？
    3. 角色（Persona）：是否指定 AI 扮演特定角色（如資訊科學老師、資料科學家、工程師、資工系教授）？
    4. 對象（Recipients）：是否交代輸出內容的目標讀者或使用者（如學生、教師、開發者）？
    5. 範本（Template）：是否提供回應格式、語言風格或其他輸出限制條件？

    步驟二：請檢查原始提問是否符合以下三項 prompt 設計原則，並對每項原則分別回答以下兩點：
    - 檢查結果：說明是否符合原則，若不符合請指出問題。
    - 補充建議：提供具體修改方向以提升該原則的符合度。

    三項設計原則如下：
    1. 簡潔（Concise）：語句是否清楚、是否去除冗詞？
    2. 有邏輯（Logical）：內容結構是否合理、順序是否清晰？
    3. 明確（Explicit）：是否明白交代所需輸出內容與格式？
    """
    # Call the chain to get the planning reasons.
    # Capture the response directly.
    Question_reasons = chain.invoke({"question": planning_prompt})

    # Step 2: 綜合整理，寫成建議
    generation_prompt = f"""這是我想到的 prompt 建議內容：
    {Question_reasons}

    以下是一些可參考的範例：
    {retrieved_chunks}

    請你根據上面的內容，輸出以下格式的內容：

    針對原本的 prompt 的建議：
    （請在這裡簡明扼要地提供修改或優化提示的建議）

    範例：
    1. （範例提問句1）
    2. （範例提問句2）
    3. （範例提問句3）
    4. （範例提問句4）
    5. （範例提問句5）

    請勿回應分析內容，只需依照上述格式輸出提問建議與範例。
    """


    # Call the chain to get the final post (suggestion).
    # Capture the response directly.
    final_post = chain.invoke({"question": generation_prompt})

    # Now, update the chat_history for the chatbot display.
    # Append the original user prompt and the combined AI response as dictionaries.
    combined_ai_response = f"🧠 AI的思考：\n{Question_reasons}\n\n💡 建議：\n{final_post}"
    chat_history.append({"role": "user", "content": prompt})
    chat_history.append({"role": "assistant", "content": combined_ai_response})

    # Return the captured values
    return Question_reasons, final_post
```

```python
def handle_click(user_question):
      # Question now returns Question_reasons and final_post
      Question_reasons, final_post = Question(user_question)
      global chat_history
      # Return Question_reasons, final_post, and the globally updated chat_history
      return Question_reasons, final_post, chat_history

with gr.Blocks() as demo:
    gr.Markdown("### 提問小助手")
    gr.Markdown("請輸入你的問題？（如：什麼是機器學習），AI會提供你問問題的建議，如果AI沒有給你範例請重試，或刷新頁面")

    with gr.Row():
        with gr.Column():
            user_input = gr.Textbox(label="你的問題是?")
            btn = gr.Button("生成建議✨")
        with gr.Column():
            chatbot = gr.Chatbot(label="歷史紀錄", value=[], type="messages")

    with gr.Row():
        out1 = gr.Textbox(label="😍AI的思考（Planning CoT）")
        out2 = gr.Textbox(label="📣建議")

    chat_history = [] # Initialize chat_history here

    btn.click(handle_click, inputs=[user_input], outputs=[out1, out2, chatbot])
```
