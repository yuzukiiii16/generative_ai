標題：應用PISA2012問題解決過程，引導使用者對LLM進行提問之AI對話機器人

製作原因：因為我的研究需要引導學生對LLM進行提問，但是擔心引導提問的過程LLM會直接給出答案，且沒辦法根據我想要的方式引導學生要如何問問題，因此我希望製作一個對話機器人，可以讓學生初步詢問要怎麼向LLM問問題，而不是只是叫LLM生成一個prompt(因為這樣不需要思考)，我希望能引導學生思考為什麼要這樣問，最後希望能提升學生問題解決能力

預計會用到：RAG、CoT

RAG：首先我會先根據PISA2012的問題解決步驟，以及我的論文整理出的prompt設計方式製作多個問題範例參考，並以markdown方式轉成向量資料餵給AI

CoT：接著我希望讓學生知道為什麼要這樣問問題，所以會用CoT模式，讓學生看到AI怎麼思考，為什麼這樣思考

也會設定system prompt讓AI不會直接和學生說答案，以及利用繁體中文回答，並引導學生思考

最後我希望這個對話機器人能夠和我的網頁結合，讓學生在使用網頁時，可以詢問我的提問機器人，最後再用這個prompt去問LLM
